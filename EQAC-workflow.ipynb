* read a row of data from the dataframe
* add a member ID, a test value, a method ID, a test ID, trial ID to a result table

```python
test_dict = {
        'albumin': {'id': 1, 'columns': [3,4,5]},
        'alp': {'id': 2, 'columns': [5,6,7]},
    }
```

```python
def insert_result(test_id, col_idx, member_id_col=0):
    '''
        columns : a list of column indexes for a test
    '''
    colnames = [data.columns[i] for i in col_idx]
    col_idx.insert(0, member_id_col)

    for rec in data[col_idx].dropna(how='all', subset=colnames).iterrows():
        _, s = rec
        print(test_id)
        print(s)
```

```python
# educational

for test in test_dict.iteritems():
    tname, tinfo = test
    insert_result(tinfo['id'], tinfo['columns'])
```
